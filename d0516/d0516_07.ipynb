```python
# 출생아수: 막대 그래프
# 합계출산율: 꺾은선 그래프
```

```python
plt.figure(figsize=(10,5)) # 그래프 배경크기 설정
num = plt.bar(df.columns,df.loc['출생아 수'])
ratio = plt.plot(df.columns,(df.loc['합계 출산율']*500),marker='o',mfc='r')
for i, txt in enumerate(df.loc['출생아 수']):
    plt.text(df.columns[i],df.loc['출생아 수'][i]+5,txt,ha='center')
for i, txt in enumerate(df.loc['합계 출산율']):
    plt.text(df.columns[i],df.loc['합계 출산율'][i]*510,txt,ha='center')
plt.yticks([300,400,500,600])
plt.ylim(250,700)
plt.legend()
```

```python
fig,ax1 = plt.subplots(figsize=(10,5))
fig.suptitle('출생아수 및 합계출산율') # 상단타이틀 출력

ax1.bar(df.columns,df.loc['출생아 수'],color='orange')
# ax1의 x축은 같이 사용할거에요
ax2 = ax1.twinx()
ax2.plot(df.columns,df.loc['합계 출산율'],marker='o',mfc='r')
ax1.set_yticks([300,400,500,600,700])
ax1.set_ylim(250,700)
ax2.set_ylim(0.5,1.5)

for i, txt in enumerate(df.loc['출생아 수']):
    ax1.text(df.columns[i],df.loc['출생아 수'][i]-15,txt,ha='center')
for i, txt in enumerate(df.loc['합계 출산율']):
    ax2.text(df.columns[i],df.loc['합계 출산율'][i]+.02,txt,ha='center')
```
