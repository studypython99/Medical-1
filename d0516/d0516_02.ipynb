```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import random
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
import pandas as pd
```

```python
data = {
   '영화' : ['명량', '극한직업', '신과함께-죄와 벌', '국제시장', '괴물', '도둑들', '7번방의 선물', '암살'],
   '개봉 연도' : [2014, 2019, 2017, 2014, 2006, 2012, 2013, 2015],
   '관객 수' : [1761, 1626, 1441, 1426, 1301, 1298, 1281, 1270], # (단위 : 만 명)
   '평점' : [8.88, 9.20, 8.73, 9.16, 8.62, 7.64, 8.83, 9.10]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
# 개봉 년도별로 출력
# 선 red, marker표시
# x축 눈금표시 2005,2010,2015,2020
# y축 눈금범위 7,10
```

```python
# T 
x = df.groupby('개봉 연도')['평점'].mean().index
y = df.groupby('개봉 연도')['평점'].mean().values
plt.plot(x,y)
```

```python
year = df.groupby('개봉 연도')['평점'].mean().index
point = df.groupby('개봉 연도')['평점'].mean().values
```

```python
plt.plot(year,point,color='r',marker='o',mec='red')
# 
for i,txt in enumerate(point):
    plt.text(year[i],point[i]+0.1,txt,ha='center',color='blue')
plt.xticks(year)
plt.ylim(7,10)
plt.grid(axis='y',ls='--',alpha=0.4)
```

```python
# 평점이 9점 이상인 영화의 비율을 확인할 수 있는 원그래프
# label: 9점 이상, 9점 미만으로 나눔
# 퍼센트: 소수점 첫째자리까지 표시
# 범례: 그래프 우측에 표시
```

```python
df.info()
```

```python
# T
filt = df['평점'] >= 9
df[filt]
len(df[filt])
# 9점이상, ~9점미만, ~(not)
values = [len(df[filt]),len(df[~filt])]
title = ['9점 이상', '9점 미만']
exp = [0.1,0]
plt.legend(loc=(1.1,0.5))
plt.pie(values,labels=title,explode=exp,autopct='%.1f%%',startangle=90,counterclock=False)
plt.title('영화 평점 그래프')
```

```python
up_9 = (df[df['평점'] >= 9])
down_9 = df[df['평점'] < 9]
```

```python
values = [up_9,down_9]
title = ['9점 이상', '9점 미만']
plt.pie(values,labels=title,autopct='%.1f%%')
```
