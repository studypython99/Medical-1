```python
# 나이가 50 이상
df['Age']>=50
len(df[df['Age']>=50])
```

```python
#상관관계 표시, 1과 가까울수록 상관관계가 크다
data_corr = df.corr()
fig = plt.figure(figsize=(8,8))
sns.heatmap(data_corr,annot=True)
```
