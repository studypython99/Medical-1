```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
import random
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
plt.bar(df['이름'],df['국어'])
```

```python
plt.plot(df['이름'],df['영어'])
plt.plot(df['이름'],df['수학'])
```

```python
# 산점도 그래프, 데이터 분포도 확인
plt.scatter(df['영어'],df['수학'])
plt.xlabel('영어점수')
plt.ylabel('수학점수')
```

```python
sizes = df['학년']*500 # 1학년*500, 2학년*500, 3학년*500
plt.scatter(df['영어'],df['수학'],s=sizes,c=df['학년'],cmap='viridis')
plt.xlabel('영어점수')
plt.ylabel('수학점수')
# y축 색상colorbar, 범위ticks, 크기shrink, 위치orientation='horizontal' 지정
plt.colorbar(ticks=[1,2,3],label='학년',shrink=0.3, orientation='horizontal')
```
