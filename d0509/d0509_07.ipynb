```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# index 기준으로 역순정렬
# ascending=True: 순차정렬, ascending=False: 역순정렬
df.sort_index(ascending=False)
```

```python
df.sort_values('키',ascending=False)
# 두가지 기준 적용, 같은키의 경우 이름으로 정렬
df.sort_values(['키','이름'],ascending=False)
# 키는 역순정렬, 이름은 순차정렬
df.sort_values(['키','이름'],ascending=[False,True])
```

```python
df = pd.read_csv('../z20_data/stu_score1.csv')

df
```

```python
#kor: 역순, id: 순차
df.sort_values(['kor','name'],ascending=[False,True])
```

```python
#평균 75점 이상이면서, 평균: 역순, 이름: 순차
avg_75 = df['avg']>=75
df[avg_75].sort_values(['avg','name'],ascending=[False,True])
```

```python
#id a가 포함된 평균: 역순, 국어: 순차
id_a = df['name'].str.lower().str.contains('a') #id a가 포함된
df[id_a].sort_values(['avg','kor'],ascending=[False,True]) #평균: 역순, 국어: 순차
```
