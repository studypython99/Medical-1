```python
import pandas as pd
```

### skiprows 상단부분을 지정한 크기만큼 제거 후 읽어오기

```python
df = pd.read_csv('stu.csv',skiprows=1) # skiprows=1 1번행 삭제, 1번행 컬럼명으로 바뀜

df
```

### nrows 지정한 개수만큼 가져오기

```python
df = pd.read_csv('stu.csv',nrows=4) # nrows=4 4줄 가져오세요, 컬럼을 제외한 4줄

df
```

```python
df.to_csv('stu2.csv',encoding='utf-8-sig')
```
