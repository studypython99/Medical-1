```python
# stu_score.csv for문을 사용하여 콘솔출력
with open('stu_score1.csv','r',encoding='utf8') as f:
    while True:
        data = f.readline()
        if data == '':    # 읽어올게 없으면, 끝까지 갔다면,
            break
        data_list = data.split(",")
        for d in data_list:
            print(d,end='\t')

# dict 타입의 list로 저장하시오
# {no:1, id:Joelynn,...}
```
