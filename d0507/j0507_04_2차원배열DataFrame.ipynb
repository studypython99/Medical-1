Data Frame 2차원 배열

```python
import pandas as pd

data = {
    '이름': ['강나래','강태원','강호림','김수찬','김재욱','박동현','박혜정','송근열'],
    '학교': ['구로고','구로고','구로고','구로고','구로고','단지고','단지고','단지고'],
    '키': [197,184,168,187,188,202,188,190],
    '국어': [90,40,80,40,15,80,55,100],
    '영어': [85,35,75,60,20,100,65,85],
    '수학': [100,50,70,70,10,95,45,90],
    '과학': [95,55,80,75,35,85,40,94],
    '사회': [85,25,75,80,10,85,35,95],
    'sw특기': ['Python','Java','Javascript','','','C+','PYTHON','c#']
}
```

```python
type(data)
```

```python
## dict 키값을 기준으로 데이터를 가져온다.
data['수학']
```

```python
data
```

```python
#dict 타입 출력
for key, value in data.items(): print(key,value)
```

```python
#dict 타입 출력, key값을 활용
for key in data.keys(): print(data[key])
```

```python
#학생 키가 180cm 이상인 학생만 출력하시오
for key in data.keys():
    if key =='키':
        # print(data[key]) # list타입
        for height in data[key]:
            if height >=180:
                print(height)
```
