```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/drinks.csv')
df.fillna('Na',inplace=True)
df
```

```python
#대륙별 갯수를 구하시오 as,eu,af
title = df['continent'].value_counts().index
values = df['continent'].value_counts().values
explode = [0.1,0,0,0,0,0]
wedgeprops = {'width':0.6, 'edgecolor':'w','linewidth':2}
plt.pie(values,labels=title, autopct='%.3f%%', startangle=90, 
        wedgeprops=wedgeprops, counterclock=False, explode=explode,
        pctdistance=0.7)
```
