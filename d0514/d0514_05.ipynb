```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False

x = [1,2,3]
y = [2,4,8]
```

```python
plt.plot(x,y,label='국어',linestyle='-.',marker='o')
plt.legend(loc='upper center')
plt.show()
```

```python
plt.figure(dpi=200) # 보여지는 그래프 전체크기 변경
plt.plot(x,y,label='국어',linestyle='-.',marker='o')
plt.legend(loc='upper center')
# plt.savefig('g3.png',dpi=3000) # 저장하기

#좌표에 데이터 2,4,8의 값을출력
#plt.text()
# x = [1,2,3], y=[2,4,8]: 데이터값
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+0.2,txt,ha='center',color='red')

plt.yticks([5,10])
plt.show()
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 키를 꺽은 선 그래프로 출력하시오
#이름, 키
# y 축 표시 170~210
```

```python
x= df['이름']
y = df['키']
plt.plot(x,y,linewidth=3,marker='o',ms=10,mec='blue',mfc='yellow',label='키')
plt.yticks([170,180,190,200,210])
for i, txt in enumerate(y):
    plt.text(x[i],y[i]+1,txt,ha='center',color='red')
plt.ylim(150,220) # y축 범위 지정
```

```python
#막대그래프
x= df['이름']
y = df['키']
#bar: 막대그래프 / barh: 옆으로 누운 막대그래프
plt.barh(x,y,label='키')
# plt.yticks([170,180,190,200,210])
# for i, txt in enumerate(y):
#     plt.text(x[i],y[i]+1,txt,ha='center',color='red')
plt.xlim(150,220) # x축 범위 지정
plt.title('키 그래프')
```
