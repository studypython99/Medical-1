```python
import pandas as pd
df = pd.read_csv('../z20_data/drinks.csv')
df
```

```python
df.groupby('continent')['beer','beer_servings'].sum()
```

```python
# 전체 평균보다 맥주소비량이 높은 대륙
# 1.전체 평균 소비량
beer_mean = df['beer_servings'].mean()
# 2. 대륙별 평균 소비량
continent_over_mean = df.groupby('continent')['beer_servings'].mean()
# 전체 평균보다 맥주소비량이 높은 대륙 맥주
continent_over_mean[continent_over_mean>=beer_mean]
```

```python
# total_litres_of_pure_alcohol 전체 평균보다 높은 대륙 출력
df['total_litres_of_pure_alcohol'].mean() # 평균 4.717098445595855
# 대륙별 합계 평균
df.groupby('continent')['total_litres_of_pure_alcohol'].mean()
```
