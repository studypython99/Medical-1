```python
# 평균 컬럼을 추가, grade컬럼에 A>=90, B>=80, C>=70, D>=60, F<60 
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
# 합계 컬럼 생성
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
df['평균'] = df['합계']/5
df
```

```python
# 함수사용, grade컬럼에 A>=90, B>=80, C>=70, D>=60, F<60 
def change(score):
    if score >= 90: return 'A'
    elif score >= 80: return 'B'
    elif score >= 70: return 'C'
    elif score >= 60: return 'D'
    else: return 'F'
```

```python
# grade컬럼에 A>=90, B>=80, C>=70, D>=60, F<60 
df['grade'] = df['평균'].apply(change)
df
```

##### 1. 컬럼을 대문자로 변경하는 방법
#####   결측치를 부여한 후 str함수를 사용해서 대문자로 변경
##### 2. 함수를 사용해서  nana데잍터 처리와 대문자로 변경하는 구문을 사용

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
```

```python
# 대문자 upper / fillna = PYTHON
# 이 컬럼의 값은 이거다;
df['sw특기'].fillna('PYTHON',inplace=True)
df['sw특기']=df['sw특기'].str.upper() # 대문자로 바꾼걸 특기컬럼에 입력
df
```

```python
# 함수 사용
def upper_change(txt):
    # null처리가 안되면 upper 적용이 안된다;;
    if pd.isnull(txt):
        return "PYTHON"
    else:
        return txt.upper()
```

```python
df['sw특기'] = df['sw특기'].apply(upper_change)
df
```
