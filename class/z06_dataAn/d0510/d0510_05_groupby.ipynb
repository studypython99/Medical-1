```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score3.csv')
df
```

```python
# 학교별로 국어점수가 가장 높은점수 출력
df.groupby(['school','grade'])['kor'].max()
```

```python
# 요약통계
df.describe()
```

```python
# 남, 여 count 출력
df.groupby('gender').count()
```

```python
#학교,성별로, 국어,영어,수학점수 평균 출력
df.groupby(['school','gender'])[['kor','eng','math']].mean().sort_values('kor',ascending=False)
```

```python
#학교,성별로, 국어,영어,수학점수 평균 출력
# df.groupby(['school','gender'])[['kor','eng','math']].mean().loc['인덱스명']
df.groupby(['school','gender'])[['kor','eng','math']].mean().loc['디지털고'] #디지털고만 보고싶다
```

```python
# agg 함수: 기본함수 호출
df.groupby('grade')['total'].agg(['mean','min','max','sum'])
```

```python
#합계컬럼 추가, 학년별 국어점수의 평균
df['total'] = df['kor']+df['eng']+df['math']
df.groupby(['school','grade'])['kor'].mean()
```
