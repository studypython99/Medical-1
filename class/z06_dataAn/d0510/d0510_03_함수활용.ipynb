```python
import random
```

```python
# 0~5까지의 함수를 랜덤으로 입력
df['요일'] = 0
for i in range(8):
    df.iloc[i,9] = random.randint(0,6)
df
```

```python
# 0=일요일, 123456=토요일
def day(dd):
    if dd==1: return '월요일'
    elif dd==2: return '화요일'
    elif dd==3: return '수요일'
    elif dd==4: return '목요일'
    elif dd==5: return '금요일'
    elif dd==6: return '토요일'
    else: return '일요일'
    
```

```python
df['요일'] =df['요일'].apply(day)
df
```

```python
# by 창민
date = ['월','화','수','목','금','토','일']
df['요일'] =df['요일'].apply(lambda x : date[x]+'요일')
df
```
