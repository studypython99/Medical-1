```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
## 조건 검색부분
lang = ['python','Java'] # 대소문자 구분
df['sw특기'].isin(lang)
```

```python
## 조건 검색부분
# 문자열함수 적용 - lower(), 소문자로 변경
# isin, 이중에 하나라도 있으면 가져오세요. 2개 이상의 검색이 가능함.
lang = ['python','java']
df['sw특기'].str.lower()
filt = df['sw특기'].str.lower().isin(lang)
df[filt]
```

```python
# java라는 글자가 포함되어 있는 것을 검색
# na =False, Nan데이터를 True or False 처리해준다.
filt = df['sw특기'].str.lower() # 문자(str)를 소문자로 바꿈 
filt = filt.str.contains('java',na=False) # 문자(str)를 포함하는
filt
```
