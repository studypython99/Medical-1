```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

##### fillna - Nan 데이터 처리: 채우기 함수

```python
# Nan을 ''(공백)으로 처리
df.fillna('')
df.fillna('없음') # 데이터 전체에서 Nan부분을 처리
df['sw특기'].fillna('없음',inplace=True)# 특정컬럼의 fillna
```

```python
df
```

```python
import numpy as np
df['sw특기'].fillna('없음',inplace=True)# 특정컬럼의 fillna
df['학교'] = np.nan
df
```

```python
df.fillna('없음')
```

##### dropna - nan처리: 지우기
##### axis: index, columns
##### how: any, all

```python
#Nan이 있는 컬럼을 전체 삭제
df.dropna(axis='columns')
```

```python
# Nan 이 있는 컬럼에서 모든 컬럼이 Nan인것만 삭제
df.dropna(axis='columns',how='all',inplace=True)
df
```

```python
# Nan 이 있는 컬럼에서 하나라도 포함되어 있다면 삭제
df.dropna(axis='columns',how='any',inplace=True)
df
```

```python
# 4번 김수찬 영어 Nan 으로 만들기
df.loc['4번','영어'] = np.NaN
df
# Nan 날리기
df.dropna(axis='columns',how='any',inplace=True)
df
```

```python
# Nan 들어가 있는 index에서 1개라도 Nan이 포함되어 있다면 삭제
df.dropna(axis='index',how='any',inplace=True)
df
```
