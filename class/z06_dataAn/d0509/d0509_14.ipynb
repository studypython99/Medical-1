```python
# 사회 컬럼을 삭제,
#   음악 컬럼 추가, 과학점수 80 이상시 음악: 80, 나머지는 65
#   합계 컬럼 추가
```

```python
df.drop(columns=['사회'],inplace=True)
df
```

```python
df['음악'] = 0
df
```

```python
df.loc[df['과학'] >=80,'음악'] = 80
df.loc[df['과학'] < 80,'음악'] = 65
df
```
