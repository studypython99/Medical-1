```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/stu_score1.csv')
df.head(3)
```

```python
h_df = df.head(20)
x = df['id'].head(20)
y = df['total'].head(20)
y1 = df['kor'].head(20)
y2 = df['eng'].head(20)
y3 = df['math'].head(20)
```

```python
plt.bar(x,y)
plt.ylim(120,300)
plt.xticks(rotation=45)
```

```python
h_df.shape[0] # 20
```

```python
idx = np.arange(h_df.shape[0])
idx
```

```python
plt.figure(figsize=(15,5))
plt.bar(idx-0.2,y1,width=0.3) # 
plt.bar(idx,y2,width=0.2) # 
for i,txt in enumerate(y2):
    plt.text(idx[i],y2[i],txt,ha='center')
plt.bar(idx+0.2,y3,width=0.3) # 
plt.xticks(idx,x,rotation=45)
```
