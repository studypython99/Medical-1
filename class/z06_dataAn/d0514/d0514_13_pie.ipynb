```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
values = [30,25,20,13,10,2] #100, 꼭 맞출 필요는 없다
# values = [1,1,1,1,1,1] # 비율로 맞춰준다
title = ['python','Java','javascript','C#','C','ETC']
explode = [0.2,0,0,0,0,0] # 0.2만큼 조각을 떼어넨다
colors = ['#3f6d91','#62617d','#4c2ad4','#613573','#d966b1','#ad0a2b']
wedgeprops={'width':0.6, 'edgecolor':'w','linewidth':2}
```

```python
plt.pie(values,labels=title,autopct='%1.f%%',colors=colors,
    wedgeprops=wedgeprops, startangle=90,counterclock=False)
```

```python
import pandas as pd
df = pd.read_csv('../z20_data/drinks.csv')
df.fillna('Na',inplace=True)
df
```

```python
#대륙별 갯수를 구하시오 as,eu,af
title = df['continent'].value_counts().index
values = df['continent'].value_counts().values
explode = [0.1,0,0,0,0,0]
plt.pie(values,labels=title, autopct='%.3f%%', startangle=90, counterclock=False, explode=explode)
```
