```python
import matplotlib.pyplot as plt# 시각화 도구 라이브러리
import matplotlib# 그래프 한글설정
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # for window
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' # for mac
matplotlib.rcParams['font.size'] = '15' # 글씨 크기
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스(-)표시
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

```python
# 이름과 국어 성적으로 선 그래프 출력
y = df['이름']
x1 = df['국어']
x2 = df['영어']
x3 = df['수학']
x4 = df['과학']
x5 = df['사회']
plt.title('빅데이터 성적그래프')

plt.bar(x1,y,label='국어성적')
plt.bar(x2,y,label='영어성적')
plt.bar(x3,y,label='수학성적')
plt.bar(x4,y,label='과학성적')
plt.bar(x5,y,label='사회성적')
plt.legend(loc=1,ncol=5)# 범례 표시, 위치설정 / ncol: 한줄에 쓰여지는 갯수

```

```python
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv',sep='\t')
df['item_price'] = df['item_price'].str.replace('$','').astype(float)
df
```

```python
# 메뉴중에서 주문을 가장 많이한 5개 @@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@
x1 = df.groupby('item_name').sum().sort_values('quantity',ascending=False).head().index
y1 = df.groupby('item_name').sum().sort_values('quantity',ascending=False).head().values

# df[df.groupby('item_name').value_counts('quantity',ascending=False)]
```

```python
df['item_name'].value_counts().head()
x2 = df['item_name'].value_counts().head().index
y2 = df['item_name'].value_counts().head().values
```

```python
#그래프 출력
plt.plot(x1,y1)
plt.plot(x2,y2)
```
