```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df
```

```python
x = df.head(5)['이름']
y = df.head(5)['국어']
y2 = df.head(5)['영어']
y3 = df.head(5)['수학']
```

```python
plt.plot(x,y,label='국어',marker='D')
plt.plot(x,y2,label='영어',marker='o',linestyle='--')
plt.plot(x,y3,label='수학',marker='s',ls='-.')
for i,txt in enumerate(y3):
    plt.text(x[i],y3[i],txt,ha='center')
# ncol = 3 // 1줄에 3개 표시
plt.legend(loc='upper center',ncol=3)
plt.show()
```

```python
colors = ['r','g','b','y','m']
plt.bar(x,y3,label='국어',color=colors)

for i,txt in enumerate(y3):
    plt.text(x[i],y3[i],txt,ha='center')
plt.xticks(rotation=45) # x축 컬럼 45도 기울이기
# ncol = 3 // 1줄에 3개 표시
plt.legend(loc='upper right')
plt.show()
```
