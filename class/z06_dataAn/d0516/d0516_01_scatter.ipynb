```python
import matplotlib.pyplot as plt
import matplotlib
import numpy as np
matplotlib.rcParams['font.family'] = 'Malgun Gothic'
matplotlib.rcParams['font.size'] = 10
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/20240514인구.xlsx', thousands=',', skiprows=3,index_col='행정기관',usecols='b,e:y')
# for i in range(df.shape[0]):
#     df.iloc[i].str.replace(',','').astype('int64[pyarrow]')
# df.head()
df.head()
```

```python
df['bot'] = df['0~4세']+df['5~9세']+df['10~14세']
df['mid'] = df['15~19세']+df['20~24세']+df['25~29세']+df['30~34세']+df['35~39세']+df['40~44세']+df['45~49세']+df['50~54세']+df['55~59세']+df['60~64세']
df['top'] = df['65~69세']+df['70~74세']+df['75~79세']+df['80~84세']+df['85~89세']+df['90~94세']+df['95~99세']+df['100세 이상']
df.head(5)
```

```python
x = df.index
y1 = df['bot']
y2 = df['mid']
y3 = df['top']
```

```python
plt.bar(x,y1,label='0~14세')
plt.bar(x,y2,label='14~60세')
plt.bar(x,y3,label='60세 이상')
plt.legend(loc='upper center')

for i, txt in enumerate(y1):
    plt.text(x[i],y1[i],txt,ha='center')
for i, txt in enumerate(y2):
    plt.text(x[i],y1[i]+y2[i],txt,ha='center')
for i, txt in enumerate(y3):
    plt.text(x[i],y1[i]+y2[i]+y3[i],txt,ha='center')
```
