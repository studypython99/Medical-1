판다스 전처리

```python
# 데이터 전처리: data, result 분리
d_target = df['Species'].to_numpy()
d_input = df.iloc[:,1:].to_numpy()
d_target
```

```python
# 데이터 전처리: train, test 분리
from sklearn.model_selection import train_test_split
train_input, test_input, train_target, test_target = train_test_split(
    d_input, d_target, random_state=42
)
```

```python
# 데이터 전처리: 표준화 작업
# 타입이 numpy 데이터만 적용 가능
from sklearn.preprocessing import StandardScaler

ss = StandardScaler()
ss.fit(train_input)
train_scaled = ss.transform(train_input)
test_scaled = ss.transform(test_input)
```

```python
# 데이터 학습시키기
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(train_scaled,train_target)
```

```python
# 정답률
print(lr.score(train_scaled,train_target)) #0.80
print(lr.score(test_scaled,test_target)) # 0.85
# train 0.50 < test 0.85 과소적합
# 반복횟수를 늘려서 정답률을 높이자
```

```python
# max_iter=1000(반복횟수), C=20(규제)를 늘려서 정답률을 높이자
lr = LogisticRegression(max_iter=1000, C=20)
lr.fit(train_scaled,train_target)
print(lr.score(train_scaled,train_target)) # 0.93
print(lr.score(test_scaled,test_target)) # 0.92
```

```python
# 경사하강법 머신러닝 튜닝, 기존 훈련된 모델에서 다시 훈련,
from sklearn.linear_model import SGDClassifier
sc = SGDClassifier(loss='log_loss',max_iter=10, random_state=42)
sc.fit(train_scaled,train_target)
```

```python
print(sc.score(train_scaled,train_target)) # 0.93
print(sc.score(test_scaled,test_target)) # 0.92
```

```python
train_target
```

```python
# partial_fit: 데이터 일부를 가져와서 target 맞추는 훈련을 한다
# 따라서, 전체적 target을 알려줘야한다.
classes = np.unique(train_target)
sc.partial_fit(train_scaled,train_target,classes=classes)
```

```python
print(sc.score(train_scaled,train_target)) # 0.93
print(sc.score(test_scaled,test_target)) # 0.92
```

```python
# partial_fit 반복, 효율의 증가
for _ in range(300):
    sc.partial_fit(train_scaled,train_target,classes=classes)
    .append(sc.score(train_scaled,test_target))
```
