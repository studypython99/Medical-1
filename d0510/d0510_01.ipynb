```python
import numpy as np
# Nan으로 변경하기 fillna
#3번 영어 /4번 수학 /5번 국어,영어 를 Nan 처리하시오.
df.loc['3번','영어'] = np.nan
df.loc['4번','수학'] = np.nan
df.loc['5번',['국어','영어']] = np.nan
df
```

```python
df.fillna(0,inplace=True)
df['국어'] = df['국어'].astype(int)
df.info()
```

```python
#총합 컬럼을 만들어서, 국영수과사 합계를 입력 및 출력
df['총합'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
df # 합계에 Nan...
```

```python
df.describe() # 음~ 대략 정보가 이렇게 되는구나 화긴★
```

```python
# 총합이 300점 이상이면 합격, 아니면 불합격
df['결과'] = '불합격' # 전부 불합격인 결과 컬럼 생성
df['총합'] >=300 # 조건
df.loc[df['총합'] >=300,'결과'] = '합격' # 결과
# 400점 이상 우수
df.loc[df['총합'] >=400,'결과'] = '우수합격' # 결과
df
```
