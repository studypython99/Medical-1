### iloc 출력, 번호로 출력

```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

```python
df.iloc[0] # df.loc['1번'] 과 같음
df.iloc[3] # 4번째 row데이터 출력
df.iloc[2:4] # 3번째 부터 4번째 row데이터 출력
df.iloc[[2,3,4],[0,2]] #3,4,5번의 이름,키
df.iloc[[4,7]]
df.loc[['5번','8번']] #df.iloc[[4,7]] 같은값
df.shape # (8, 9)
```

```python
#4번 ㅎ학생의 키 출력
df.loc['4번','키'] #187
df.iloc[3,2] #187

#4번, 6번 학생의 국어, 사회 출력
df.loc[['4번','6번'],['국어','사회']]
df.iloc[[3,-3],[3,-2]]

df.loc['3번':'4번']
df.iloc[2:4]

df.loc['3번':'4번','이름':'영어']
# df.iloc[2:4,0:5]
```
